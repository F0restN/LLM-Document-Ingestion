```python
import pandas as pd
import json

# Read the JSON file
with open('aging-care-forum.json', 'r') as file:
    data = json.load(file)

# Convert to DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
df = df.rename(columns={'question_context': 'context'})
df = df.rename(columns={'answer': 'content'})
```

```python
df.head()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
import pandas as pd

# 创建文本分割器
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

# 初始化存储列表
document_chunks = []

# 处理DataFrame中的每一行
for doc in df.itertuples():
    # 只处理content字段
    content = str(doc.content) if doc.content is not None else ""
    
    # 只有当content不为空时才进行分块
    if content.strip():
        for chunk in text_splitter.split_text(content):
            document_chunks.append({
                'content': chunk,
                'url': doc.url,
                'title': doc.title,
                'question': doc.question,
                'context': doc.context,
                'tags': doc.tags,
                'source-tag': "agingcare"
            })

# 转换为DataFrame
chunks_df = pd.DataFrame(document_chunks)

# 保存为CSV
chunks_df.to_csv('aging-care-forum-chunked.csv', index=False)

# 打印信息
print(f"Total number of chunks created: {len(chunks_df)}")
print("\nFirst few rows of the chunked data:")
print(chunks_df.head())
```

```python
import pandas as pd
import json

# Read the JSON file
with open('pubmed-central-delirium-family-caregiving.json', 'r') as file:
    data = json.load(file)

# Convert to DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
df = df.rename(columns={'contents': 'content'})
df.head()
```

```python
df['context'] = ''
df.head()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
import pandas as pd

# 创建文本分割器
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

# 初始化存储列表
document_chunks = []

# 处理DataFrame中的每一行
for doc in df.itertuples():
    # 只处理content字段
    content = str(doc.content) if doc.content is not None else ""
    
    # 只有当content不为空时才进行分块
    if content.strip():
        for chunk in text_splitter.split_text(content):
            document_chunks.append({
                'content': chunk,
                'doi': doc.doi,
                'pmid': doc.pmid,
                'authors': doc.authors,
                'abstract': doc.abstract,
                'per_sections_content': doc.per_sections_content,
                'published_date': doc.published_date,
                'reference': doc.reference,
                'url': doc.url,
                'title': doc.title,
                'context': doc.context,
                'source-tag': "pubmed"
            })

# 转换为DataFrame
chunks_df = pd.DataFrame(document_chunks)

# 保存为CSV
chunks_df.to_csv('pubmed-central-delirium-family-caregiving-chunks.csv', index=False)

# 打印信息
print(f"Total number of chunks created: {len(chunks_df)}")
print("\nFirst few rows of the chunked data:")
print(chunks_df.head())
```
